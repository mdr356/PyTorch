```python
# Relu Activation function is function tha returns X if x is greater than 0 otherwise it returns 0
## Relu introduces non-linearity into a neural network, without non-linearity,
## a neural network would only be a linear transfromation. 
import torch
import matplotlib.pyplot as plt


def Relu(x: torch.Tensor) -> torch.Tensor:
    return torch.max(torch.Tensor([0]), x)
```

```python
#example
A = torch.arange(start=-10, end=10, step=1, dtype=torch.float32)
plt.subplot(2,2,1)
plt.plot(A)
plt.title('Before Relu')

print(A)
A_relu = Relu(A)
plt.subplot(2,2,2)
plt.plot(A_relu)
plt.title("After Relu function")
plt.show()
print(f"\n\nAfter Relu: \n{A_relu}")
```
